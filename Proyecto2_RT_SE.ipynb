#Preprocesamiento

```python
df = df.drop('customer_id', axis=1)
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis') #Visualización de valores nulos
plt.show()

# Remueve valores nulos
df = df.dropna()
```

```python
X = df.drop('credit_score', axis=1) #dataset sin la variable a predecir
y = df['credit_score'] #Variable a predecir
```

```python

X['interest_occupation_rank'] = X.groupby('occupation')['interest_rate'].rank() #Se crea la columna interest_ occupation

X['age_bin'] = pd.cut(X['age'], bins=[0, 18, 35, 50, 100], labels=[1, 2, 3, 4]) #Se discretiza la edad por rangos

bins = [0, 30000, 50000, 80000, 120000, float('inf')]
labels = [1, 2, 3, 4, 5]

X=X.drop('occupation', axis=1)
# Create a new column 'income_bin' based on the bins


X['income_bin'] = pd.cut(X['annual_income'], bins=bins, labels=labels, right=False) #Se discretiza el ingreso
X=X.drop('annual_income', axis=1)


bins = [0, 5000, 10000, 20000, 50000, float('inf')]

X['monthly_inhand_salary_category'] = pd.cut(X['monthly_inhand_salary'], bins=bins, labels=labels) #se discretiza el salario mensual en mano


X=X.drop('monthly_inhand_salary', axis=1)
categories = ['Low_spent_Small_value_payments',
              'High_spent_Small_value_payments',
              'Low_spent_Medium_value_payments',
              'High_spent_Medium_value_payments',
              'Low_spent_Large_value_payments',
              'High_spent_Large_value_payments']



# Se mapean las categorias a ordinales
category_mapping = {category: code for code, category in enumerate(categories)}

# Se mapea la columna categorica a ordinal
X['payment_behaviour_ordinal'] = X['payment_behaviour'].map(category_mapping)
X=X.drop('payment_behaviour', axis=1)

```

# Pipelines

```python
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# Se realiza pipeline para las variables numericas
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Se realiza pipeline para las variables categoricas

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

# Se crea el columnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

preprocessor.set_output(transform="pandas") #Tiene salida en formato pandas


```

Como se observa en la sección de preprocesamiento, los datos nulos en el pipeline se imputan. Sin embargo, este procedimiento se realiza después de haber eliminado las filas con datos nulos en la instancia no preprocesada de los datos. Esto se lleva a cabo de esta manera porque, en el proceso de crear columnas en el feature engineering, se generan columnas que en ciertas instancias producen valores nulos. Estos valores, preferentemente, son imputados, ya que estas filas contienen información valiosa presente en las otras columnas del conjunto de datos.

Primero, se crea la columna 'interest_occupation_rank' mediante el uso de la función rank(). Esta columna asigna un rango a las tasas de interés ('interest_rate') según la ocupación ('occupation') a la que pertenecen, proporcionando así una medida relativa de las tasas de interés dentro de cada grupo ocupacional.

Luego, se procede a discretizar la variable 'age' en rangos específicos utilizando la función pd.cut(). La nueva columna 'age_bin' asigna etiquetas numéricas a diferentes grupos de edades, lo que puede ser útil para capturar patrones específicos relacionados con la edad en el análisis posterior.

El siguiente paso implica la discretización del ingreso anual ('annual_income') mediante la creación de la columna 'income_bin'. Esta transformación permite agrupar los ingresos en categorías predefinidas, simplificando la variabilidad y posiblemente destacando patrones o tendencias asociadas con diferentes niveles de ingresos.

A continuación, se realiza una operación similar para el salario mensual en mano ('monthly_inhand_salary'). La columna 'monthly_inhand_salary_category' se crea a través de la discretización, dividiendo los salarios en categorías definidas por intervalos específicos. Esta transformación puede facilitar la identificación de relaciones no lineales entre el salario y otras variables.

Finalmente, se aborda la variable categórica 'payment_behaviour'. Se realiza un mapeo de las categorías a valores ordinales mediante un diccionario de mapeo ('category_mapping'). ## Optimización Random Forest

```python
def objective_RF(trial):
    # Definir los parámetros a optimizar
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_depth = trial.suggest_int('max_depth', 5, 30)
    min_samples_split = trial.suggest_float('min_samples_split', 0.1, 1.0)
    min_samples_leaf = trial.suggest_float('min_samples_leaf', 0.1, 0.5)

    # Crear el modelo RandomForestClassifier con los parámetros sugeridos
    model_RF = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )

    # Combinar preprocesamiento y modelo en un único pipeline
    pipeline_RF = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model_RF)
    ])

    # Entrenar el modelo en el conjunto de entrenamiento
    pipeline_RF.fit(X_train, y_train)

    # Predicciones en el conjunto de validación
    y_pred_val = pipeline_RF.predict(X_val)

    # Calcular la métrica a optimizar (precisión en este caso) en el conjunto de validación
    accuracy_val = accuracy_score(y_val, y_pred_val)

    return accuracy_val

# Crear un estudio Optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective_RF, n_trials=100)



# Visualización de resultados de Optuna
optuna.visualization.plot_optimization_history(study)
optuna.visualization.plot_slice(study)

# Obtener el mejor modelo con los parámetros encontrados
best_model_RF = RandomForestClassifier(
    n_estimators=study.best_params['n_estimators'],
    max_depth=study.best_params['max_depth'],
    min_samples_split=study.best_params['min_samples_split'],
    min_samples_leaf=study.best_params['min_samples_leaf'],
    random_state=42
)

# Combinar preprocesamiento y el mejor modelo en un único pipeline
best_pipeline_RF = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', best_model_RF)
])

# Entrenar el mejor modelo en todo el conjunto de entrenamiento
best_pipeline_RF.fit(X_train, y_train)

# Predicciones en el conjunto de prueba
y_pred_test = best_pipeline_RF.predict(X_test)

# Calcular y mostrar la métrica en el conjunto de prueba (precisión en este caso)
test_accuracy = accuracy_score(y_test, y_pred_test)
print("Precisión en el conjunto de prueba:", test_accuracy)

# Predicciones en el conjunto de validación
y_pred_val = best_pipeline_RF.predict(X_val)

# Calcular y mostrar la métrica en el conjunto de validación (precisión en este caso)
val_accuracy = accuracy_score(y_val, y_pred_val)
print("Precisión en el conjunto de validación:", val_accuracy)

# Imbalanced Learn - Aplicar Random Over-sampling para tratar el desequilibrio
ros_RF = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros_RF.fit_resample(X_train, y_train)

# Volver a entrenar el modelo con el conjunto de entrenamiento resampleado
best_pipeline_RF.fit(X_train_resampled, y_train_resampled)

# Predicciones en el conjunto de prueba después del resampling
y_pred_test_resampled = best_pipeline_RF.predict(X_test)

# Calcular y mostrar la métrica en el conjunto de prueba después del resampling (precisión en este caso)
test_accuracy_resampled = accuracy_score(y_test, y_pred_test_resampled)
print("Precisión en el conjunto de prueba después del resampling:", test_accuracy_resampled)

```

## Optimización SVM

```python
def objective_SVM(trial):
    # Define SVM parameters to optimize
    C = trial.suggest_loguniform('C', 1e-3, 1e3)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly', 'sigmoid'])
    degree = trial.suggest_int('degree', 2, 5)
    gamma = trial.suggest_categorical('gamma', ['scale', 'auto'])

    # Create SVM model with suggested parameters
    model_SVM = SVC(C=C, kernel=kernel, degree=degree, gamma=gamma, random_state=42)

    # Combine preprocessing and model into a single pipeline
    pipeline_SVM = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model_SVM)
    ])

    # Train the model on the training set
    pipeline_SVM.fit(X_train, y_train)

    # Predictions on the validation set
    y_pred_val = pipeline_SVM.predict(X_val)

    # Calculate the metric to optimize (accuracy in this case) on the validation set
    accuracy_val = accuracy_score(y_val, y_pred_val)

    return accuracy_val

# Create an Optuna study
study = optuna.create_study(direction='maximize')
study.optimize(objective_SVM, n_trials=10)

# Visualization of Optuna results
optuna.visualization.plot_optimization_history(study)
optuna.visualization.plot_slice(study)

# Get the best SVM model with the found parameters
best_model_SVC = SVC(
    C=study.best_params['C'],
    kernel=study.best_params['kernel'],
    degree=study.best_params['degree'],
    gamma=study.best_params['gamma'],
    random_state=42
)

# Combine preprocessing and the best model into a single pipeline
best_pipeline_SVM = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', best_model_SVC)
])

# Entrenar el mejor modelo en todo el conjunto de entrenamiento
best_pipeline_SVM.fit(X_train, y_train)

# Predicciones en el conjunto de prueba
y_pred_test = best_pipeline_SVM.predict(X_test)

# Calcular y mostrar la métrica en el conjunto de prueba (precisión en este caso)
test_accuracy = accuracy_score(y_test, y_pred_test)
print("Precisión en el conjunto de prueba:", test_accuracy)

# Predicciones en el conjunto de validación
y_pred_val = best_pipeline_SVM.predict(X_val)

# Calcular y mostrar la métrica en el conjunto de validación (precisión en este caso)
val_accuracy = accuracy_score(y_val, y_pred_val)
print("Precisión en el conjunto de validación:", val_accuracy)

# Imbalanced Learn - Aplicar Random Over-sampling para tratar el desequilibrio
ros_SVM = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros_SVM.fit_resample(X_train, y_train)

# Volver a entrenar el modelo con el conjunto de entrenamiento resampleado
best_pipeline_SVM.fit(X_train_resampled, y_train_resampled)

# Predicciones en el conjunto de prueba después del resampling
y_pred_test_resampled = best_pipeline_SVM.predict(X_test)

# Calcular y mostrar la métrica en el conjunto de prueba después del resampling (precisión en este caso)
test_accuracy_resampled = accuracy_score(y_test, y_pred_test_resampled)
print("Precisión en el conjunto de prueba después del resampling:", test_accuracy_resampled)

```

#Interpretabilidad

**Modelo Propuesto**

Para evaluar la interpretabilidad de un modelo de machine learning y poder explicar cómo toma decisiones basadas en los datos, es crucial emplear diversas técnicas que ofrezcan una comprensión clara y precisa. Entre las herramientas más efectivas se encuentran la determinación de la importancia de las características y el uso de SHAP (SHapley Additive exPlanations).

La evaluación de la importancia de las características permite identificar qué variables tienen una mayor influencia en las decisiones del modelo. Este enfoque proporciona una visión general de cómo el modelo utiliza diferentes entradas para realizar predicciones. Sin embargo, es importante considerar las limitaciones inherentes a este método, como su sensibilidad a la multicolinealidad y la posible omisión de interacciones complejas entre características.

Por otro lado, SHAP, una técnica basada en la teoría de juegos, ofrece una explicación más detallada al asignar contribuciones específicas de cada característica a la predicción del modelo. SHAP proporciona una descomposición aditiva que revela cómo cada característica afecta la salida del modelo, teniendo en cuenta la interacción con otras variables. Esto facilita una comprensión más profunda de las decisiones del modelo, incluso en casos de relaciones no lineales o dependencias complejas.

Es por el motivo anterior que se logra un análisis comprehensivo y robusto de la interpretabilidad del modelo al combinar ambas tecnicas. Las importancias de características brindan una visión general, SHAP ofrece explicaciones detalladas para predicciones individuales, y PDP permite explorar las relaciones entre variables, complementando así la comprensión del comportamiento del modelo en diferentes niveles de detalle.
